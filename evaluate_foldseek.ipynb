```python
from pathlib import Path
import pandas as pd
```

```python
cluster_tsv = pd.read_csv(Path("test_outputs_C6/aa_info_cluster.tsv"), sep="\t", header=None)
```

```python
len(cluster_tsv[0].unique())
```

```python
# count the number samples per cluster
cluster_counts = cluster_tsv[0].value_counts().reset_index()
print(cluster_counts)
```

```python
# Sort files according to cluster size for download

cluster_dir = Path("test_outputs_C6/res_clusters_aa/")
cluster_dir.mkdir(exist_ok=True, parents=True)

filepath = Path("test_outputs_C6/output_homomers_filtered")

# Iterate over all clusters and copy to cluster directories
for i, (cluster, count) in enumerate(cluster_counts.itertuples(index=False)):
    files = cluster_tsv[cluster_tsv[0] == cluster].index.tolist()
    print(f"Cluster {cluster} has {count} samples, files: {files}")
    cluster_path = cluster_dir / Path(f"cluster_{i:02d}_{count:04d}")
    cluster_path.mkdir(exist_ok=True, parents=True)
    for file in files:
        file_name = cluster_tsv.iloc[file][1] + ".pdb"
        src = filepath / file_name
        dst = cluster_path / file_name

        # copy file if it exists
        if src.exists():
            dst.write_text(src.read_text())
```
